# 3. Model Comparison

```python
import matplotlib.pyplot as plt
import pandas as pd

# Model performance data
models = ["Logistic Regression", "Random Forest", "K-Nearest Neighbors", "Support Vector Machine", "Decision Tree"]
accuracy = [0.9474, 0.9649, 0.9649, 0.9825, 0.9474]
precision = [0.9577, 0.9589, 0.9718, 0.9726, 0.9577]
recall = [0.9577, 0.9859, 0.9718, 1.0000, 0.9577]

# Calculate F1-Score
f1 = [2 * (precision[i] * recall[i]) / (precision[i] + recall[i]) for i in range(len(models))]

# Create DataFrame and sort by Accuracy
data = {
    "Model": models,
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1-Score": f1
}

df = pd.DataFrame(data)
df = df.sort_values(by="Accuracy", ascending=False).reset_index(drop=True)

# Print table
print("\nModel Comparison Summary:")
print(df)

# Best and worst models
best_model = df.iloc[0]
worst_model = df.iloc[-1]

# Print best and worst model insights
print("\nBest Performing Algorithm:")
print(f"{best_model['Model']} performed the best with an accuracy of {best_model['Accuracy']:.4f} "
      f"and F1-Score of {best_model['F1-Score']:.4f}.")

print("\nWorst Performing Algorithm:")
print(f"{worst_model['Model']} performed the worst with an accuracy of {worst_model['Accuracy']:.4f} "
      f"and F1-Score of {worst_model['F1-Score']:.4f}, indicating possible overfitting or poor generalization.")

# Visualization
bar_width = 0.2
index = range(len(models))

plt.figure(figsize=(12, 6))
plt.bar([i - 1.5 * bar_width for i in index], accuracy, width=bar_width, label='Accuracy')
plt.bar([i - 0.5 * bar_width for i in index], precision, width=bar_width, label='Precision')
plt.bar([i + 0.5 * bar_width for i in index], recall, width=bar_width, label='Recall')
plt.bar([i + 1.5 * bar_width for i in index], f1, width=bar_width, label='F1-Score')

plt.xlabel('Classification Models')
plt.ylabel('Score')
plt.title('Performance Comparison of Classification Algorithms')
plt.xticks(index, models, rotation=30)
plt.ylim(0.85, 1.0)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
